```python

import matplotlib.pyplot as plt

# Calculating the percentage of 0s and 1s
zero_percentage = (df['Caudal_impulses'] == 0).mean() * 100
one_percentage = (df['Caudal_impulses'] == 1).mean() * 100

# Plotting
plt.figure(figsize=(6, 4))
plt.bar(['0', '1'], [zero_percentage, one_percentage], color=['red', 'blue'])
plt.title('Percentage of 0s and 1s in Caudal Impulses')
plt.xlabel('Caudal Impulses')
plt.ylabel('Percentage')
plt.ylim(0, 100) 
plt.show()

```

```python
#creating a pie cart
labels = ['0', '1']
sizes = [zero_percentage, one_percentage]
colors = ['red', 'blue']
explode = (0.1, 0) 
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of 0s and 1s in Caudal Impulses')
plt.axis('equal')  
```
