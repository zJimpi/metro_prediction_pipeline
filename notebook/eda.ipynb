```python

import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'Caudal_impulses' column

# Calculate the percentage of 0s and 1s
zero_percentage = (df['Caudal_impulses'] == 0).mean() * 100
one_percentage = (df['Caudal_impulses'] == 1).mean() * 100

# Plotting
plt.figure(figsize=(6, 4))
plt.bar(['0', '1'], [zero_percentage, one_percentage], color=['red', 'blue'])
plt.title('Percentage of 0s and 1s in Caudal Impulses')
plt.xlabel('Caudal Impulses')
plt.ylabel('Percentage')
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100
plt.show()

```
